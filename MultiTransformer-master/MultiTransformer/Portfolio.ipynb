```python
import yfinance as yf
import pandas as pd
# Define the ticker list
indices = ['AAPL', 'MSFT', 'NVDA', 'JNJ', 'NVS','JPM','GS','AMZN','DIS','MCD','NEE','BA','CAT','XOM','CVX','RIO','BHP']

# Create an empty DataFrame
data = pd.DataFrame()

# Get data on each ticker
for index in indices:
    indexData = yf.Ticker(index)
    indexDf = indexData.history(period='1d', start='2000-1-1', end='2008-12-31')
    
    # add this data to the main DataFrame
    data[index] = indexDf['Close']

# See your data

```
