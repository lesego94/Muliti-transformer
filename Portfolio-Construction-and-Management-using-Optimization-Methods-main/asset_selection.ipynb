```python
import pandas as pd
import getFamaFrenchFactors as gff
```

```python
file1 = pd.DataFrame()
file_list = ["BCOMTR.xlsx","HFRIFWI.xlsx","LBUSTRUU.xlsx","LG30TRUU.xlsx","LMBITR.xlsx","NDDUE15X.xlsx","NDDUJN.xlsx",
            "NDDUNA.xlsx","NDDUPXJ.xlsx","NDDUUK.xlsx","NDDUWXUS.xlsx","NDUEEGF.xlsx","RU10GRTR.xlsx","RU10VATR.xlsx",
            "RU20GRTR.xlsx","RU20INTR.xlsx","RU20VATR.xlsx","RUMCINTR Equityl.xlsx","RUMRINTR Equity.xlsx",
            "RUTPINTR Equity.xlsx","S5COND Equity.xlsx","S5CONS Equity.xlsx","S5ENRS Equity.xlsx",
            "S5FINL Equity.xlsx","S5HLTH Equity.xlsx","S5INDU Equity.xlsx","S5INFT Equity.xlsx",
            "S5MATR Equity.xlsx","S5RLST Equity.xlsx","S5TELS Equity.xlsx","S5UTIL Equity.xlsx",
            "SPXT US Equity.xlsx"]
for i in file_list:
    file = pd.read_excel(i,engine='openpyxl').iloc[7:,:2]
    file.set_index("Security",inplace=True)
    file.sort_index(inplace=True)
    file= file.resample('M').last()
    file1=file1.join(file,how="outer")
```

```python
file2 = file1.drop(columns = ['NDDUNA Index                                                    ',
                              'NDDUPXJ Index                                                   ',
                              'S5RLST Index                                                    ',
                              'SPXT US Equity                                                  ',
                              'NDUEEGF Index                                                   ',
                              'RUMRINTR Index                                                  '])
file3 = file2.dropna()
file3
```

```python
monthly_returns = file3.pct_change()
```

```python
ff3_monthly = gff.famaFrench3Factor(frequency='m')
ff3_monthly.rename(columns={"date_ff_factors": 'Date'}, inplace=True)
ff3_monthly.set_index('Date', inplace=True)
rf = ff3_monthly['RF']
rf = pd.DataFrame(rf)
data = rf.join(monthly_returns,how='right')
data
```

```python
data.isnull().sum().sum()
```

```python
description = data.describe() 
new = description.T
new
```

### sort by sharpe ratio

```python
new["sharpe_ratio"] = ((new["mean"]-new["mean"]['RF'])/ new["std"])
sharpe_high=new.sort_values(by=['sharpe_ratio']).tail(15)
sharpe_high
```

```python
sharpe_filter=sharpe_high.index
sharpe_filter
```

### sort by mean of return

```python
new_by_return=new.sort_values(by=['mean'])
new_by_return.loc[sharpe_filter].sort_values(by=['mean'])
```

### sort by volatility

```python
new_by_volatility=new.sort_values(by=['std'])
new_by_volatility=new_by_volatility.loc[sharpe_filter].sort_values(by=['std'])
new_by_volatility
```

```python
low_risk=new_by_volatility['std'].head(5).index
low_risk
```

```python
high_risk=new_by_volatility['std'].tail(5).index
high_risk
```

```python
total_asset=low_risk.append(high_risk)
total_asset
```

```python
data = file3[total_asset]
data
```

```python
data.to_excel('data.xlsx', sheet_name='data', index=True)
```
